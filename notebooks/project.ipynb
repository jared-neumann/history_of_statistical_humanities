## Display basic graphs

```python
# Convert 'PERIODS_OF_10_YEARS' to string for better plotting
df_table_1['PERIODS_OF_10_YEARS'] = df_table_1['PERIODS_OF_10_YEARS'].astype(str)

# Plotting the data
plt.figure(figsize=(12, 8))

# Plot for each category
plt.plot(df_table_1['PERIODS_OF_10_YEARS'], df_table_1['COLLECTED_WORKS'], label='Collected Works', marker='o')
plt.plot(df_table_1['PERIODS_OF_10_YEARS'], df_table_1['SEPARATE_PLAYS_OR_POEMS'], label='Separate Plays/Poems', marker='o')
plt.plot(df_table_1['PERIODS_OF_10_YEARS'], df_table_1['ALTERED_PLAYS'], label='Altered Plays', marker='o')
plt.plot(df_table_1['PERIODS_OF_10_YEARS'], df_table_1['SPURIOUS_PLAYS'], label='Spurious Plays', marker='o')
plt.plot(df_table_1['PERIODS_OF_10_YEARS'], df_table_1['COMMENTARIES'], label='Commentaries', marker='o')
plt.plot(df_table_1['PERIODS_OF_10_YEARS'], df_table_1['TOTAL_PUBLICATIONS'], label='Total Publications', marker='o', linestyle='--', color='black')

# Adding titles and labels
plt.title('Shakespearean Literature Publications Over Decades')
plt.xlabel('Periods (Decades)')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
